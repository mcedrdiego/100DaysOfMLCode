```python
import numpy as np 
import pandas as pd 
import seaborn as sns

```

```python
url = "http://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"

df = pd.read_csv(url, delimiter=',', header=None)
df.columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'class']
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
df = df.dropna(subset=['petal_width']) # Not necessary coz no missing values
df.info()
```

```python
sns.pairplot(df, hue='class', height=2.5)
```

```python
final_df = df[df['class'] != 'Iris-virginica']
sns.pairplot(final_df, hue='class')
final_df.head()
```

```python
final_df.hist(column = 'petal_length', bins=20, figsize=(10,5))
```
